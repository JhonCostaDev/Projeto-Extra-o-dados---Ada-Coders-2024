```python
import requests
import os 
```

```python


# importando as variáveis de ambiente do .env 

newsapi = os.getenv('NEWS_API_KEY')

url = f'https://newsapi.org/v2/everything?q=apple&from=2024-10-21&to=2024-10-21&sortBy=popularity&apiKey={newsapi}&language=en'

response = requests.get(url)
response
```

```python
articles_json = response.json()['articles']
```

```python
for i in range(len(articles_json)):
    print(articles_json[i].get('title'))

```

```python
import pandas as pd 
df = pd.DataFrame(articles_json)
df
```
